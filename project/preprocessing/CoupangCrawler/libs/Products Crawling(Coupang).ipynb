```python
# 카테고리별 사이트 주소 크롤링
from requests import get, post
from bs4 import BeautifulSoup
import re

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Whale/3.23.214.10 Safari/537.36', "Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}
resp = get('https://www.coupang.com/', headers = headers)
dom = BeautifulSoup(resp.text, 'html5lib')

categories = list() # 쿠팡 본 사이트에서 추출할 수 있는 카테고리들의 url 수집
for a in dom.find_all('a'):
    try:
        href = a.attrs['href']
        if re.match(r'/np/categories', href):
            categories.append('https://www.coupang.com'+ href)
    except:
        pass

# 대분류 카테고리 안 소분류 카테고리 url 수집(디테일한 부분에서 조금 완성도가 떨어지는 것 같지만, 많은 데이터 수집 가능)
res = set() # 최종 url 결과
for url in categories:
    res.add(url)
    resp = get(url, headers = headers)
    dom = BeautifulSoup(resp.text, 'html5lib')
    try:
        for i in dom.select('map > area'):
            href = i.attrs['href']
            if re.match('/np/categories', href):
                res.add('https://www.coupang.com' + href[:21]) 
    except:
        pass
```

```python
# 소분류 카테고리 하나에 대한 {상품명:[가격, 이미지 url, 사이트 url]} 딕셔너리 저장
import os
from urllib.request import urlretrieve, urlopen

url = 'https://www.coupang.com/np/categories/144042'
rst = dict() # 상품 정보 저장 딕셔너리
for i in range(1, 18):
    params = {'page':i}
    resp = get(url, headers = headers, params = params)
    dom = BeautifulSoup(resp.text, 'html5lib')
    
    for a in dom.select('a[class="baby-product-link"]'):
        link = 'https://www.coupang.com' + a.attrs['href']
        img = 'https:' + a.select_one('dt > img').attrs['src'] # 일단은 대문 이미지 하나만 가져옴(추가 필요시 수정 가능)
        name = a.select_one('dd > div[class="name"]').text.strip()
        price = a.select_one('strong[class="price-value"]').text
        rst[name] = [price, img, link]

# 상품 이미지 파일 저장(.jpg)
os.mkdir('practice')
folder_name = 'practice'
for k, v in rst.items():
    img_link = v[1]
    try:
        name = re.sub(r'[\/:*?"<>|]', ' ', k) # 파일 저장 시 금지 문자 치환
        urlretrieve(img_link, f'{folder_name}/{name}.jpg') # 이미지 저장
    except:
        name = k
        urlretrieve(img_link, f'{folder_name}/{name}.jpg') 
```
